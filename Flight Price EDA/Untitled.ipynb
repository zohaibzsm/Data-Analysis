## Handling datetime columns

```python
# handling Date_of_Journey
final_df["Date_of_Journey_Day"] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df["Date_of_Journey_Month"] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df["Date_of_Journey_Year"] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
# changing type to integer
final_df["Date_of_Journey_Day"] = final_df['Date_of_Journey_Day'].astype(int)
final_df["Date_of_Journey_Month"] = final_df['Date_of_Journey_Month'].astype(int)
final_df["Date_of_Journey_Year"] = final_df['Date_of_Journey_Year'].astype('int')
```

```python
final_df.info()
```

```python
# removing this column as this is useless now
final_df.drop('Date_of_Journey', axis=1, inplace=True)
final_df.head()
```

```python
# handling Arrival_Time
final_df['Arrival_Time_Hour'] = final_df.Arrival_Time.str.split(' ').str[0].str.split(':').str[0]
final_df['Arrival_Time_Minute'] = final_df.Arrival_Time.str.split(' ').str[0].str.split(':').str[1]
```

```python
# changing type to integer
final_df["Arrival_Time_Hour"] = final_df['Arrival_Time_Hour'].astype(int)
final_df["Arrival_Time_Minute"] = final_df['Arrival_Time_Minute'].astype(int)
```

```python
# handling Dep_Time
final_df['Dep_Time_Hour'] = final_df.Dep_Time.str.split(' ').str[0].str.split(':').str[0]
final_df['Dep_Time_Minute'] = final_df.Dep_Time.str.split(' ').str[0].str.split(':').str[1]
```

```python
# changing type to integer
final_df["Dep_Time_Hour"] = final_df['Dep_Time_Hour'].astype(int)
final_df["Dep_Time_Minute"] = final_df['Dep_Time_Minute'].astype(int)
```

```python
final_df.info()
```

```python
# removing the necessary columns 
final_df.drop(columns=['Arrival_Time', 'Dep_Time'], inplace=True)
```
